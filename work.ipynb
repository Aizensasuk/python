```python
#int to other 
a=100 #int value
b=float(a)#covert int into float
c=bool(a)#int into boolean 
d=complex(a)#int into complex
e=str(a)#int into string
#type of every thing
print(a,type(a))
print(b,type(b))
print(c,type(c))
print(d,type(d))
print(e,type(e))
```

```python
#float into other 
a=100.33#float value
b=int(a)#covert  float into int
c=bool(a)#float into boolean 
d=complex(a)#float into complex
e=str(a)#float into string
#type of every thing
print(a,type(a))
print(b,type(b))
print(c,type(c))
print(d,type(d))
print(e,type(e))
```

```python
#complex into other 
a=100+2j#complex value
b=float(a.real)#covert complex into float
c=bool(a)#complex into boolean 
d=int(a.real)#complex into int
e=str(a)#complex into string
#type of every thing
print(a,type(a))
print(b,type(b))
print(c,type(c))
print(d,type(d))
print(e,type(e))
```
